````python
##########################地铁路线规划问题
#######地铁数据获取
import requests
from bs4 import BeautifulSoup
import re
import string
from urllib.parse import quote

##伪装为浏览器
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36',}
url = 'https://baike.baidu.com/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'##北京地铁图
response = requests.get(url,headers=headers)
response = response.content.decode("UTF-8")

##爬取地铁线的百度百科网址,获取地铁线路和对应url
bs = BeautifulSoup(response,"html.parser")
urls_html = bs.find("caption",text="北京地铁开通简表").parent

urls_dict = {}
pattern = re.compile('<a href="(/item/.+?)" target="_blank">([\w\u4e00-\u9fa5]+线)</a>')
for (url,name) in re.findall(pattern,str(urls_html)):
    urls_dict[name]= "https://baike.baidu.com" + url
list(urls_dict)
urls_dict
````

````python
####根据urls_dict中的地铁线路爬取地铁站
lines_stations = {}
for (name,url) in list(urls_dict.items()):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36',}
    response = requests.get(url,headers=headers)
    response = response.content.decode("UTF-8").replace("br","").replace("br/","")##先去除掉标签<br>，不然下面使用bs提取标签文本的时候，有时候会返回None
    bs = BeautifulSoup(response,"html.parser")
    
    if bs.find("b",text=re.compile(".*首末.*车[(时刻时间)]+.*")) is  not None:
        table_html = bs.find("b",text=re.compile(".*首末.*车[(时刻时间)]+.*")).parent.parent.parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].find(re.compile("(td)|(th)")).get_text()
            if len(s)<9 and s not in ["车站名称","首车时间","备注"]:
                stations.append(s)
        #print(name,stations)
        lines_stations[name]=stations
    elif bs.find("caption",text=re.compile(".*首末.*车[(时刻时间)]+.*")) is not None:
        table_html = bs.find("caption",text=re.compile(".*首末.*车[(时刻时间)]+.*")).parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].find(re.compile("(td)|(th)")).get_text()
            if len(s)<9 and s not in ["车站名称","首车时间","备注"]:
                stations.append(s)
        #print(name,stations)
        lines_stations[name]=stations
    elif bs.find("td",text=re.compile(".*首末.*车[(时刻时间)]+.*")) is not None:
        table_html = bs.find("td",text=re.compile(".*首末.*车[(时刻时间)]+.*")).parent.parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].find(re.compile("(td)|(th)")).get_text()
            if len(s)<9 and s not in ["车站名称","首车时间","备注"]:
                stations.append(s)
        #print(name,stations)
        lines_stations[name]=stations
    elif bs.find("caption",text=re.compile("车站信息.*")) is not None:
        table_html = bs.find("caption",text=re.compile("车站信息.*")).parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].findAll(re.compile("(th)|(td)"))[1].get_text().replace("\n","")
            if len(s)<9 and s not in ["车站名称","首车时间"]:
                stations.append(s)
        #print(name,stations)
        lines_stations[name]=stations
    elif bs.find("th",text=re.compile("车站名称")) is not None:
        table_html = bs.find("th",text=re.compile("车站名称")).parent.parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].find(re.compile("(div)|(th)"))
            if s is not None:
                if len(s.get_text())<9 and s.get_text() != "车站名称": 
                    stations.append(s.get_text())
        #print(name,stations)
        lines_stations[name]=stations
    elif bs.find("div",text=re.compile("车站名称")) is not None:
        table_html = bs.find("div",text=re.compile("车站名称")).parent.parent.parent
        stations = []
        for i in range(len(table_html.findAll("tr"))):
            s = table_html.findAll("tr")[i].find("div")
            if s is not None:
                if len(s.get_text())<9 and s.get_text() != "车站名称":            
                    stations.append(s.get_text())
        #print(name,stations)
        lines_stations[name]=stations
    else:
        print(name,url)


lines_stations.pop("复八线")
lines_stations
````

````python
############地铁站点数据预处理
from collections import defaultdict
##去掉空值,去掉换行符号，统一站名，在每个站后面添加站字
for (line,stations)in lines_stations.items():
    stations = [str(s).replace("\n","") for s in stations  if s !="" and s not in ("1611","全程","编号","车站名","时间" )]
    stations = [s+"站" if not re.findall(re.compile(".*站"),s) else s for s in stations]
    #print(line,set(stations))
    stations_set = list(set(stations))
    stations_set.sort(key=stations.index)
    lines_stations[line]=stations_set
    
with open(r"C:\Users\zhangxing\Documents\GitHub\NLP-COURSE\Assignments\Lesson-02\subway_stations_data.txt","w",encoding="utf-8") as f:
    for (name,station) in lines_stations.items():
        f.write(name)
        f.write(":")
        f.write(" ".join(station))
        f.write("\n")    

    

##将站点名之间的连接关系用字典表示出来
stations_map = {}
stations_map = defaultdict(list)
for name,s in lines_stations.items():
    #循环每一个站，把每个站点的上一个站和下一个站加入到字典中
    stations_map[s[0]].append(s[1])
    stations_map[s[1]].append(s[0])
    for i in range(1,len(s)-1):
        previous = s[i-1]
        latter = s[i+1]
        stations_map[s[i]].append(previous)
        stations_map[s[i]].append(latter)
    stations_map[s[len(s)-1]].append(s[len(s)-2])
for name,s in stations_map.items():
    stations_map[name] = list(set(s))


````

````python
##########定义地铁路线搜索函数，输入起点和终点，可以输出地铁路路线
##广度优先遍历
def search(start,destination):
    pathes = [[start]]
    visited = set()#记录拓展过的地铁站点
    
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]#有待拓展的地铁站点
        if frontier not in  visited:#如果frontier没有拓展过，就执行下面语句，继续拓展frontier，successor为frontier要拓展为的站点
            successor = stations_map[frontier]
            for city in successor:
                if city in path:continue #避免环
                new_path  = path + [city]
                pathes.append(new_path)
                #print(new_path)
                if city == destination:return " -> ".join(new_path)
            visited.add(frontier)

search("莲花桥站","知春里站")
````

````python
##上面的代码问题：两两站点之间搜索路线，之后再拼接。可能在最后拼接的路线出现了环
##########定义不同的搜索策略,路线中途经过某个指定站点by_way
def search_two_station(start,destination):
    pathes = [[start]]
    visited = set()#记录拓展过的地铁站点
    
    while pathes:
        path = pathes.pop(0)
        frontier = path[-1]#有待拓展的地铁站点
        if frontier not in  visited:#如果frontier没有拓展过，就执行下面语句，继续拓展frontier，successor为frontier要拓展为的站点
            successor = stations_map[frontier]
            for city in successor:
                if city in path:continue #避免环
                new_path  = path + [city]
                pathes.append(new_path)
                #print(new_path)
                if city == destination:return new_path
            visited.add(frontier)
        pathes = sorted(pathes,key=len)
        #pathes = sorted(pathes,key=len,reverse=True)
        
def search(start,destination,by_way):#从start站到destination站，途径by_way站点
    all_stations = [start] + by_way
    pathes = [start]
    
    while len(all_stations)>=2:
        temp_start = all_stations.pop(0)
        temp_destination = all_stations[0]
        new_pathes = search_two_station(temp_start,temp_destination)[1:]
        for i in new_pathes:
            if i in all_stations: all_stations.remove(i) 
        pathes.extend(new_pathes)
    pathes.extend(search_two_station(pathes[-1],destination)[1:])
    return " -> ".join(pathes)

search("莲花桥站","中关村站",["苹果园站","北京西站"])


````

````python
##更友好的输出，添加地铁线路图
def pretty_print(stations):
    new_print = []
    stations = stations.split(" -> ")
    for i in range(len(stations)-1):
        for name,lines in lines_stations.items():
            if set([stations[i],stations[i+1]]) < set(lines) : new_print.append("=======".join([stations[i],name]))
    new_print.append(stations[-1])
    return " =======> ".join(new_print)

pretty_print(search("莲花桥站","中关村站",["苹果园站","北京西站"]))
````
